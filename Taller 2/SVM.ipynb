# Importando librerias necesarias

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn import metrics

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Cargando datos

```python
data_train = pd.read_csv('train.csv')
data_test = pd.read_csv('test.csv')
pd.set_option('display.max_columns', None) 
data_train.describe()
```

- Los tipos de suelo 7 y 15 son constantes, pueden ser borrados
- Las escalas de las variables no son las mismas. Por lo tanto, deben ser normalizadas.
- Los id, no son información valiosa para construir un modelo

```python
data_train = data_train.drop(['Soil_Type7', 'Soil_Type15'], axis = 1)
data_test = data_test.drop(['Soil_Type7', 'Soil_Type15'], axis = 1)

data_train = data_train.iloc[:,1:]
data_test = data_test.iloc[:,1:]
```

# Normalizando

```python
x = data_train[data_train.columns.drop('Cover_Type')].values
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
data_train_features = pd.DataFrame(x_scaled, columns=data_train.columns.drop('Cover_Type')) 
data_train_features
```

# Correlaciones

```python
size = 10
data = data_train.iloc[:,:size]

cols = data.columns

data_corr = data.corr()

# Threshold: se tendran en cuentas las variables con correlación alta
threshold = 0.5
corr_list = []
data_corr
```

```python
# Filtrando correlaciones altas
for i in range(0, size):
    for j in range(i+1, size):
        if data_corr.iloc[i,j] >= threshold and data_corr.iloc[i,j]<1 \
        or data_corr.iloc[i,j] < 0 and data_corr.iloc[i,j] <= -threshold:
            corr_list.append([data_corr.iloc[i,j],i,j])
            
# Ordenando valores de correlaciones
s_corr_list = sorted(corr_list,key= lambda x: -abs(x[0]))

# Imprimiendo correlaciones más altas
for v,i,j in s_corr_list:
    print("%s vs %s = %.2f" % (cols[i], cols[j], v))            
```

# Skewness

```python
data_train.iloc[:,:10].skew()
```

```python
columns_ind = ['Elevation', 'Aspect', 'Slope', 'Horizontal_Distance_To_Hydrology', 'Vertical_Distance_To_Hydrology', 'Horizontal_Distance_To_Roadways',
               'Hillshade_9am', 'Hillshade_Noon', 'Hillshade_3pm', 'Horizontal_Distance_To_Fire_Points']
```

# SVM

```python
train, test = train_test_split(data_train, test_size=0.2, random_state=1)
train_features = train[columns_ind].values
test_features = test[columns_ind].values
train_covers = train['Cover_Type']
test_covers = test['Cover_Type']
```

```python
svm = SVC(gamma='scale', decision_function_shape='ovo', kernel='rbf', C=100)
svm.fit(train_features, train_covers)
```

```python
svm.score(train_features, train_covers)
```

```python
predicted_covers = svm.predict(test_features)
report = metrics.classification_report(y_true=test_covers, 
                                           y_pred=predicted_covers, 
                                           labels=np.unique(predicted_covers))
print(report)
```
